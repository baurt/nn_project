```python
import torch
import torchvision
import torch.nn as nn
from torch.utils.data import DataLoader
from torchvision import datasets
from torchvision import transforms as T
from torchvision import io
import torchutils as tu
import json
import numpy as np
import matplotlib.pyplot as plt
import os
```

```python
import os

# Путь к каталогу с данными
data_dir = '/input/cub2002011/CUB_200_2011'

# Вывести список файлов и подкаталогов в указанном каталоге
print(os.listdir(data_dir))
```

```python
## Определите аугментации для обучающих данных
train_transform = T.Compose([
    T.Resize((224, 224)),
    T.RandomHorizontalFlip(),
    T.ToTensor(),
])

# Определите преобразования для валидационных данных (масштабирование)
val_transform = T.Compose([
    T.Resize((224, 224)),
    T.ToTensor(),
])

# Создайте ImageFolder для обучающих и валидационных данных
train_dataset = torchvision.datasets.ImageFolder(
    os.path.join(data_dir, 'images'),  # Полный путь к каталогу 'train'
    transform=train_transform
)

train_loader = DataLoader(train_dataset, batch_size=32, shuffle=True)

# Создайте DataLoader для обучающих и валидационных данных
class_names = train_dataset.classes

plt.figure(figsize=(15, 25))
batch_size = 32
# for i in range(batch_size):
#     plt.subplot(7, 5, i+1)
#     plt.axis('off')
#     plt.imshow(images[i].permute(1, 2, 0))
#     plt.title(class_names[labels[i].item()])
```

```python
train_size = int(0.8 * len(train_dataset))
valid_size = len(train_dataset) - train_size
train_data, valid_data = torch.utils.data.random_split(train_dataset, [train_size, valid_size])
```

```python
class_names
```

```python
valid_loader = DataLoader(valid_data, batch_size=32, shuffle=True)
```

```python
import torchvision.models as models
```

```python
model = models.resnet18(pretrained=True)
```

```python
model.to(device)
```

```python
num_ftrs = model.fc.in_features
model.fc = nn.Linear(num_ftrs, 200)
```

```python
model = model.to("cuda")
```

```python
tu.get_model_summary(model, torch.randn(32, 3, 224, 224, device="cuda"))
```

```python
for param in model.parameters():
    param.requires_grad = False
model.fc.weight.requires_grad = True
model.fc.bias.requires_grad = True

for param in model.layer4.parameters():
    param.requires_grad = True
```

```python
optimizer = torch.optim.Adam(model.parameters())
criterion =  torch.nn.CrossEntropyLoss()
```

```python
for param in model.parameters():
  print(param.requires_grad)
```

```python
# Обучение модели
num_epochs = 5
for epoch in range(num_epochs):
    model.train()
    running_loss = 0.0
    for inputs, labels in train_loader:
        inputs, labels = inputs.to(device), labels.to(device)

        optimizer.zero_grad()

        outputs = model(inputs)
        loss = criterion(outputs, labels)

        loss.backward()
        optimizer.step()

        running_loss += loss.item()

    # Оценка точности модели на валидационном наборе данных
    model.eval()
    correct = 0
    total = 0
    with torch.no_grad():
        for inputs, labels in valid_loader:
            inputs, labels = inputs.to(device), labels.to(device)
            outputs = model(inputs)
            _, predicted = torch.max(outputs, 1)
            total += labels.size(0)
            correct += (predicted == labels).sum().item()

    accuracy = correct / total
    epoch_loss = running_loss / len(train_loader)
    print(f'Epoch {epoch+1}/{num_epochs}, Loss: {epoch_loss:.4f}, Accuracy: {accuracy:.2f}')
```
